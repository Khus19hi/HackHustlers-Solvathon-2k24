```python
import numpy as np
import pandas as pd
```

```python
headers=["symboling","normalized-losses","make","fuel-type","aspiration","num-of-doors","body-style","drive-wheels","engine location","wheel-base","length","width","height","curb-weight","engine-type","num-of-cylinders","engine-size",
"fuel-system","bore","stroke","compression-ratio","horsepower","peak-rpm","city-mpg","highway-mpg","price"]
ds="set.data"
df=pd.read_csv(ds,sep=",",names=headers)
```

```python
df.replace("?",np.nan,inplace=True)
```

```python
#Replace with Mean
avg_norm_loss=df["normalized-losses"].astype("float").mean(axis=0)
df["normalized-losses"].replace(np.nan,avg_norm_loss,inplace=True)
```

```python
#Replace with Mean
avg_norm_loss=df["bore"].astype("float").mean(axis=0)
df["bore"].replace(np.nan,avg_norm_loss,inplace=True)
```
